```python
customers = df[df['inbound'] == True][['tweet_id', 'text']]
customers.head()
```

```python
agents = df[df['inbound'] == False][['in_response_to_tweet_id', 'text']]
agents.head()
```

```python
faq_df = customers.merge(
    agents,
    left_on='tweet_id',
    right_on='in_response_to_tweet_id',
    suffixes=('_customer', '_agent')
)

faq_df.head()
```

```python
faq_df = faq_df[['text_customer', 'text_agent']]
faq_df.columns = ['question', 'answer']
faq_df.head()
```

```python
faq_df.dropna(inplace=True)
faq_df.drop_duplicates(inplace=True)
len(faq_df)
```

```python
import pickle

with open('../models/faq_data.pkl', 'wb') as f:
    pickle.dump(faq_df, f)
```
